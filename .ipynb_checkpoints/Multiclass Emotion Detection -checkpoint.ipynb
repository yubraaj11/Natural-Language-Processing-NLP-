```python
import pandas as pd
```

```python
data = pd.read_csv('ISEAR.csv', encoding='ISO-8859-1', names=['emotion', 'sentence'])
```

```python
data
```

```python
import re
import nltk
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lemmatizer = WordNetLemmatizer()
corpus=[]
```

```python
for i in range(0, len(data['sentence'])):
    review = re.sub('[^a-zA-Z]', ' ', data['sentence'][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('English'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(corpus).toarray()
```

```python
X
```

```python
emotion_map = {'anger':0, 'disgust':1, 'fear':2, 'guilt':3, "joy":4, 'sadness':5, 'shame':6}
data['emotion'] = data['emotion'].map(emotion_map)

# y=pd.get_dummies(data['emotion'])
# y=y.iloc[:,1].values
```

```python
y=data['emotion']
```
